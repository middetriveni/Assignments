```python
# removing the duplicates
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
# keeping the items purchased in each transaction in a list
df=df.apply(lambda x:x.str.split(','))
df.head()
```

```python
store = [row[0] for row in df.values] # Iterate through rows using df.values and access first element
```

```python
# Import the TransactionEncoder class
from mlxtend.preprocessing import TransactionEncoder # Import TransactionEncoder from mlxtend.preprocessing
te = TransactionEncoder()
arry = te.fit_transform(store)
```

```python
arry[1]
```

```python
df=pd.DataFrame(arry,columns=te.columns_)
```

```python
df.head()
```

```python
# selecting the frequent items in the transactions with minimum support = 0.009
frequent_itemsets = apriori(df, min_support=0.009, use_colnames=True)
frequent_itemsets
```

```python
# Generate association rules from the frequent itemsets
from mlxtend.frequent_patterns import association_rules # Make sure association_rules is imported
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2) # Generate rules with confidence >= 0.2
# mineral water & spaghetti are consequent for every antecedent items
rules # Display the rules you generated
```

# Interview Questions

### 1. What is Lift and why is it important in Association Rules?

Lift measures how much more often two items (A and B) occur together than if they were completely independent.
#### Formula:      
        Lift(AB)= Support(A)Support(B)/Support(AB)

#### Interpretation:    
        Lift = 1  A and B are independent (no association).

     Lift > 1  A and B appear together more often than expected (positive association).

     Lift < 1  A and B appear together less often than expected (negative association).

#### Importance:   
    It helps validate if the rule is useful or just random.

    High lift values indicate strong, meaningful associations that could be actionable in real-world decisions
    (like product recommendations).

### 2. What is Support and Confidence? How do you calculate them?

### Support:
      How frequently an itemset appears in the dataset.

##### Formula:
     Support(A) = Number of transactions containing A / Total number of transactions

##### Purpose:
     Helps filter out rarely occurring patterns; focuses on common patterns.

### Confidence:
     How often the rule has been found to be true. (Given A happened, how often did B also happen?)

##### Formula:
     Confidence(AB)= Support(A) / Support(AB)

##### Purpose:
     Measures the reliability of the inference made by a rule.

3. What are some limitations or challenges of Association Rules Mining?

#### Too many rules:
     It can generate a huge number of rules, many of which may be uninteresting or redundant.

#### Rare but important associations may be missed:
     If you set high support thresholds, you might miss rare but important relationships.

#### Interpretability:
     Its hard to make sense of thousands of rules manually; prioritizing rules needs more metrics (like lift, conviction).

#### Scalability:
     Mining association rules in large datasets can be very computationally intensive.

#### Non-causal relationships:
     Association doesn't imply causation; a rule may show a pattern but not a cause-effect relationship.
