### Dimensionality Reduction with PCA ####

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
wine_norm = scale(wine)
wine_norm
```

```python
pca =PCA()
pca_values=pca.fit_transform(wine_norm)
```

```python
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
pca.components_
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0]
y = pca_values[:,1]
#z = pca_values[:2:3]
plt.scatter(x,y)
```

```python
pca_data=pd.DataFrame(columns=["X","Y"])
pca_data.X=x
pca_data.Y=y
```

```python
pca_data
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=6).fit(pca_data)
pca_data.plot(x="X",y="Y",c=model.labels_,kind="scatter",s=50,cmap=plt.cm.coolwarm_r)
```

```python
import sklearn
sklearn.metrics.silhouette_score(pca_data,model.labels_)
```

```python
## CLUSTERING WITH ORIGINAL DATA
fig=plt.figure(figsize=(8,4))
WCSS=[]
for i in range(1,11): 
    model=KMeans(n_clusters=i).fit(wine_norm)
    WCSS.append(model.inertia_)
```

```python
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model1=KMeans(n_clusters=6).fit(wine_norm)
y_means=model1.predict(wine_norm)
y_means
```

### 2.Practical Implications of using PCA and Clustering ##

### PCA(principal component analsis ) ###

```python
Simplifies complex datasets by reducing dimensionality, making them easier to visualize
and interpret.Helps to remove noise and redundant features, improving the performance
of machine learning models.Ideal for data compression up computations in large datasets.
```

### Clustering ##

```python
Enables unsupervised groping of data, revealing hidden patterns or segments (like
customer segmentation, wine types etc.) useful for market analysis recommendation systems,
fraud detection, and more.                                                                        
```

## 3.Recommendations for using PCA and Clustering ##

### use PCA when: ###

```python
The dataset has many features(high dimensionality).we need to visualize
complex data in 2D or 3D.We want to remove noise and focus on important features.
we aim to improve model speed and performance.
```

### use clustering when: ###

```python
We want to find hidden groups or patterns in unlabeled data.
we need to perform market segmentation or classification without labels.
we plan to apply clustering after PCA for better accuracy and efficiency
```

###  Combine PCA + Clustering when: ###

```python
we want to enhance clustering performance on high-dimensional data.we need 
faster computation and more meaningful clustering, we aim to reduce reduancy
and highligh
```
